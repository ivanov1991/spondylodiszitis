```python
## LESE DIE BEVÖLERUNGSDATEN EIN.

# Erstellung eines DataFrames, in dem die Variable Altersgruppe mit den Merkmalsausprägungen unter60/über60 geführt wird 
columns=[]

for age_group in ['bis unter 60 Jahre','60 Jahre und älter']:
    
    for gender in ["männlich", "weiblich"]:

        for state in states:
            
            columns.append((age_group,gender,state))

multi_index=pd.MultiIndex.from_tuples(columns)

multi_index.names = ['Altersgruppe', 'Geschlecht', "Bundesland"]

population_dataframe_60=pd.DataFrame(columns=multi_index,index=years)
population_dataframe_60.index.name="Jahr"


for year in years:

    file_path=os.getcwd()+os.sep+"data"+os.sep+"destatis"+os.sep+"population"+os.sep+str(year)+os.sep+"12411-0013-DLAND.xlsx"
    
    population=pd.read_excel(file_path, skiprows=3,index_col=0)
    
    block_size=3

    for state in states:
        
        # Find the start index of each state in the dataframe
        state_index=list(population.columns).index(state)
        
        result=population.iloc[2:93,state_index:(state_index+block_size)]
        
        # sum 
        for group in ['bis unter 60 Jahre','60 Jahre und älter']:
            
            female_sum_over_groups=0
            male_sum_over_groups=0
            
            for destatis_group in age_group_mapping_60[group]:
                
                # index 1 is female
                female_sum_over_groups= female_sum_over_groups + result.loc[destatis_group][1]
                # index 0 is male
                male_sum_over_groups= male_sum_over_groups + result.loc[destatis_group][0]
                
            population_dataframe_60.loc[year,(group,'männlich',state)]= male_sum_over_groups
            population_dataframe_60.loc[year,(group,'weiblich',state)]= female_sum_over_groups
            
```

```python
# Erstellung eines DataFrames, in dem die Variable Altersgruppe mit allen Altersgruppen geführt wird.

columns=[]

for age_group in [*age_group_mapping_v1.keys()]:
    
    for gender in ["männlich", "weiblich"]:

        for state in states:
            
            columns.append((age_group,gender,state))

multi_index=pd.MultiIndex.from_tuples(columns)

multi_index.names = ['Altersgruppe', 'Geschlecht',"Bundesland"]

population_dataframe_all_age_groups=pd.DataFrame(columns=multi_index,index=years)
population_dataframe_all_age_groups.index.name="Jahr"


for year in years:

    file_path=os.getcwd()+os.sep+"data"+os.sep+"destatis"+os.sep+"population"+os.sep+str(year)+os.sep+"12411-0013-DLAND.xlsx"
    
    population=pd.read_excel(file_path, skiprows=3,index_col=0)
    
    block_size=3

    for state in states:
        
        # Find the start index of each state in the dataframe
        state_index=list(population.columns).index(state)
        
        result=population.iloc[2:93,state_index:(state_index+block_size)]
        
        # sum 
        for group in [*age_group_mapping_v1.keys()]:
            
            female_sum_over_groups=0
            male_sum_over_groups=0
            
            for destatis_group in age_group_mapping_v1[group]:
                
                # index 1 is female
                female_sum_over_groups= female_sum_over_groups + result.loc[destatis_group][1]
                # index 0 is male
                male_sum_over_groups= male_sum_over_groups + result.loc[destatis_group][0]
                
            population_dataframe_all_age_groups.loc[year,(group,'männlich',state)]= male_sum_over_groups
            population_dataframe_all_age_groups.loc[year,(group,'weiblich',state)]= female_sum_over_groups
            
```

```python
##### VORARBEIT #####
#### Lese die Destatis Daten ein. Das statitische Bundesamt hat eine Sonderauswertung bereitgestellt mit den fünf-stelligen Diagnose
#### Codes. 
###################

### The destatis provided a special data evaluation which provides the case numbers down to the five digit diagnoses code

# Dies ist die Hautpdiagnose so wie sie im Excel sheet genannt ist
diagnosen= ["M4620","M4621","M4622","M4623","M4624","M4625","M4626","M4627","M4628","M4629",
            "M4630","M4631","M4632","M4633","M4634","M4635","M4636","M4637","M4638","M4639",
            "M4640","M4641","M4642","M4643","M4644","M4645","M4646","M4647","M4648","M4649"]

## Initialize dataframe

columns=[]


for diagnose in diagnosen:

    for age_group in [*age_group_to_column_mapping.keys()]:
    
    
        for gender in ["männlich", "weiblich"]:

            for state in states:
            
                columns.append((diagnose, age_group, gender, state))

multi_index=pd.MultiIndex.from_tuples(columns)
multi_index.names = ['Diagnose', 'Altersgruppe', 'Geschlecht',"Bundesland"]


## Aggregate the columns to just two age groups (below 60, and 60 and beyond)

icd_dataframe_all_codes_all_age_groups=pd.DataFrame(columns=multi_index,index=years)
icd_dataframe_all_codes_all_age_groups.index.name="Jahr"


for year in years:
    
    print(year)

    ## Read in the destatis data, for all icd codes

    for age_group in [*age_group_to_column_mapping.keys()]:

        column_name=age_group_to_column_mapping[age_group]

        file_path=os.getcwd()+os.sep+"Spondylodiszitis_AG_GE_Wohnort_2005-2021_M46_2_M46_3_M46_4.xlsx"

        yearly_data=pd.read_excel(file_path, skiprows=11,usecols="A:C,"+column_name, sheet_name=str(year))
            
        for diagnose in diagnosen:

            for state in states:

                for gender in ["m","w"]:

                    cases_df=yearly_data[(yearly_data['Geschlecht']==gender) & (yearly_data["Hauptdiagnose"]==diagnose) & (yearly_data["Wohnort der Patienten (Land)"]==state)]

                    cases=cases_df[cases_df.columns[-1]].to_numpy()[0]

                    # when no cases are present
                    if cases=="-" or cases==".":

                        case_number=0

                    else:
                        case_number=cases

                    if gender=="m":

                        icd_dataframe_all_codes_all_age_groups.loc[year,(diagnose,age_group,'männlich',state)]=case_number

                    else:

                        icd_dataframe_all_codes_all_age_groups.loc[year,(diagnose,age_group,'weiblich',state)]=case_number
```

```python
###################################################################################################################
### Abbildung 6 #########################
###################################################################################################################
## Versuche hier die Inzidenz der tuberkulösen Spondylodiszitis zu berechnen

cases_tuberculosis_spondylodiszitis=[]

for year in years:

    file_path=os.getcwd()+os.sep+"A180_M462_M463_M464_2005-2021_curated.xlsx"
    
    sheet_data=pd.read_excel(file_path, sheet_name=str(year))     

    yearly_cases=0

    for index in sheet_data.index:
    
        cell=sheet_data['insgesamt'].loc[index]
    
        if (cell is not '-') and (cell is not np.nan) and (cell is not '.'):
            yearly_cases=yearly_cases+cell
            
    cases_tuberculosis_spondylodiszitis.append(yearly_cases)
    
```

```python
###############################################################################################################
## Abbildung 7 ######
##############################################################################################################

## Das Dictionary hält die Inzidenz pro Hauptdiagnose (M46.20 - M46.49)

## Abbildung 7, Plotte nach Subdiagnosen ausgeschlüsselt

## Alle Spalten, die zu einer Subdiagnose gehören werden aus dem dataframe herausgenommen, addiert und durch die
## Gesamtbevölkerung geteilt.

incidence_dict={}


for hauptdiagnose in diagnosen:
    
    inzidence_pro_diagnose=icd_dataframe_all_codes_all_age_groups.loc[:,(hauptdiagnose, slice(None), slice(None))].sum(axis=1)/population_dataframe_all_age_groups.sum(axis=1)*1e6
    incidence_dict[hauptdiagnose]=inzidence_pro_diagnose

styles=[
        'k',
        'dimgrey',
        'darkgrey',
        'rosybrown',
        'brown',
        'darkred',
        'r',
        'coral',
        'sienna',
        'sandybrown',
        'darkorange',
        'tan',
        'darkgoldenrod',
        'gold',
        'darkkhaki',
        'olive',
        'yellow',
        'greenyellow',
        'forestgreen',
        'darkslategrey',
        'powderblue',
        'dodgerblue',
        'royalblue',
        'b',
        'indigo',
        'violet',
        'magenta',
        'mediumvioletred',
        'crimson',
        'lightpink'
]


plot_array=[]

for ind,key in enumerate([*incidence_dict.keys()]):
    plot_array.append(incidence_dict[key].to_numpy())


fig, ax=plt.subplots(1,1, figsize=(20, 10))
plt.subplots_adjust(left=None, bottom=0.2, right=None, top=0.8, wspace=1, hspace=1)


plt.xlabel('Jahr', size=20)

ax.set_ylim([0,180])
ax.set_xlim([2004,2028])
ax.set_ylabel('Inzidenz [Fälle pro Jahr pro 1.000.000 Einwohner]',size=20)
ax.tick_params(labelsize=20)

ax.stackplot(years,plot_array)
            

ax.set_xticks(years)
ax.set_xticklabels(years, rotation=90)

ax.legend(incidence_dict.keys(),loc='upper right',fontsize=15, ncol=2)

fig_name = "Abbildung_7.png"
fig.savefig(fig_name)

## Schreibe die Daten in eine Excel Datei

writer = pd.ExcelWriter("Abbildung_7.xlsx", engine = 'xlsxwriter')


for ind,key in enumerate([*incidence_dict.keys()]):
    incidence_dict[key].to_excel(writer, sheet_name = key)

writer.close()
```
